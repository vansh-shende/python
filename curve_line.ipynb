```python
#write a python program to fit a curve
import numpy as np
import matplotlib.pyplot as plt
### x and y values ###

x = np.array([1,2,3,4,6])
y = np.array([0.5,2,4.5,8,12.5])


#fit a curve y=ax^b log(y) = log(a)+log(b)  y=bx+a


#convert to log log lineriation
log_x=np.log(x)
log_y=np.log(y)


# fit a line y = ax+b using numpy's polyfit

b,log_a=np.polyfit(log_x,log_y,1)
a=np.exp(log_a)

#copute fited value
x_fit=np.linspace(min(x),max(x),200)
y_fit=a*(x_fit**b)

#plot original scatter points
plt.scatter(x,y,color='red',label="sample data")

#plot best.fit line
plt.plot(x_fit,y_fit, color='blue', label=f"Best fit:y={a:.2f}x^{b:2f}")
plt.xlabel('x-axis"')
plt.ylabel('y-axis"')
plt.title("line fitting using least square")
plt.legend()
plt.grid()
plt.show()
```
